### 1. Simple Plotting

Create a plot of the function $y=x^3$ for $x\in[-10,10]$ using numpy arrays and matplotlib. Add appropriate axis labels and a title.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)
y = x**3

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y = x^3')
plt.title('Plot of y = x^3')
plt.show()
````

### 2. Line Plots

Plot $y=2x + 1$ and $y=x^2$ on the same plot as above for $x\in[-5,5]$ using different colours and linestyles. Add a legend.

````python
x = np.linspace(-5, 5, 200)
y1 = 2 * x + 1
y2 = x**2

plt.plot(x, y1, label='y = 2x + 1', color='blue', linestyle='-')
plt.plot(x, y2, label='y = x^2', color='red', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plots of y = 2x + 1 and y = x^2')
plt.legend()
plt.show()
````

### 5. Subplots

Create a `2 x 2` subplot layout and plot:

- $\sin(x)$

- $\cos(x)$

- $\tan(x)$

- $\sin(x)\cos(x)$

with $x\in[0, 2\pi]$. Label each plot.

````python
x = np.linspace(0, 2 * np.pi, 100)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# sin(x)
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('sin(x)')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('sin(x)')

# cos(x)
axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('cos(x)')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('cos(x)')

# tan(x)
axs[1, 0].plot(x, np.tan(x))
axs[1, 0].set_title('tan(x)')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('tan(x)')
axs[1, 0].set_ylim(-10, 10)  # Limit y-axis for better view

# sin(x)*cos(x)
axs[1, 1].plot(x, np.sin(x) * np.cos(x))
axs[1, 1].set_title('sin(x) * cos(x)')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('sin(x)*cos(x)')

plt.tight_layout()
plt.show()
````

#### Simulating a Planet
Copy your function `simulate_orbit` from question 4 of practical session 5, and use it to plot the orbit of a planet, replacing the mathmatical functions with `numpy`functions. Place a cross at `(0,0)`.

````python
def simulate_orbit(
    r,                     # Orbital radius (m)
    M,                     # Mass of central star (kg)
    steps=100,             # Number of steps in simulation
    G=6.67430e-11,         # Gravitational constant (default value)
    return_period=True     # Whether to return orbital period
):
    """
    Simulate the x, y positions of a planet in circular orbit using numpy.

    Parameters:
        r (float): Orbital radius in meters.
        M (float): Mass of the star in kg.
        steps (int, optional): Number of time steps. Default is 100.
        G (float, optional): Gravitational constant. Default is 6.67430e-11.
        return_period (bool, optional): Whether to return orbital period.

    Returns:
        list of tuple: x, y positions at each time step.
        float (optional): Orbital period in seconds.
    """
    v = np.sqrt(G * M / r)         # Orbital velocity
    T = 2 * np.pi * r / v          # Orbital period
    omega = 2 * np.pi / T          # Angular velocity

    dt = T / steps
    times = np.linspace(0, T, steps)
    x = r * np.cos(omega * times)
    y = r * np.sin(omega * times)

    positions = list(zip(x, y))

    if return_period:
        return positions, T
    else:
        return positions

# Example usage (Earth-Sun)
positions, period = simulate_orbit(1.496e11, 1.989e30)
print(f"Orbital period (seconds): {period:.2e}")
print(f"Orbital period (days): {period / (3600*24):.2f}")
print("First 5 positions:")
for pos in positions[:5]:
    print(pos)

# Extract x and y for plotting
x_vals, y_vals = zip(*positions)

plt.figure(figsize=(6,6))
plt.plot(x_vals, y_vals, label='Orbit Path')
plt.scatter(0, 0, marker='x', color='red', s=100, label='Star (0,0)')

plt.title('Simulated Circular Orbit')
plt.xlabel('x position (m)')
plt.ylabel('y position (m)')
plt.axis('equal')  # Equal aspect ratio for proper circle
plt.legend(loc = 'upper right')
plt.show()
````

#### Estimating $\pi$

From exercise 7 of session 7, take your `estimate_pi` function, and alter the code so it plots, the box, the unit circle, and each random point where those inside the circle are red and those outside the circle are blue. Add a legend and grid lines.

````python
def estimate_pi(n):
    # Generate n random points in [-1,1] x [-1,1]
    x = np.random.uniform(-1, 1, n)
    y = np.random.uniform(-1, 1, n)
    
    # Calculate distance from origin
    dist = np.sqrt(x**2 + y**2)
    
    # Boolean mask for points inside the unit circle
    inside_mask = dist <= 1
    
    # Count points inside the unit circle
    inside_circle = np.sum(inside_mask)
    
    # Estimate pi using ratio of points inside circle vs total points
    pi_estimate = 4 * inside_circle / n
    
    # Plotting
    fig, ax = plt.subplots(figsize=(6,6))
    
    # Plot square boundary
    square = plt.Rectangle((-1, -1), 2, 2, fill=False, linewidth=2, label='Square [-1,1] x [-1,1]')
    ax.add_patch(square)
    
    # Plot unit circle boundary
    circle = plt.Circle((0, 0), 1, fill=False, color='green', linewidth=2, label='Unit Circle (r=1)')
    ax.add_patch(circle)
    
    # Plot points inside circle in red
    ax.scatter(x[inside_mask], y[inside_mask], color='red', s=10, label='Inside Circle')
    
    # Plot points outside circle in blue
    ax.scatter(x[~inside_mask], y[~inside_mask], color='blue', s=10, label='Outside Circle')
    
    ax.set_aspect('equal')
    ax.set_xlim(-1.1, 1.1)
    ax.set_ylim(-1.1, 1.1)
    ax.set_title(f'Estimating  with n={n}\n  {pi_estimate:.5f}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True)
    ax.legend()
    
    plt.show()
    
    return pi_estimate

# Example usage
estimate_pi(500)
````
