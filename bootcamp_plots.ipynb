```python
x = [1, 2, 3, 4, 5]
y1 = [i**2 for i in x]
y3 = [5, 10, 5, 10, 5]

plt.figure(figsize=(6, 4))
plt.plot(x, y1, label='Squared', ls = '--', c='blue', lw = 4)
plt.scatter(x, y2, label='Oscillating', 
            s=300, marker = '*',c='peru', alpha=0.7)

plt.xlabel('X', fontsize = 12)
plt.ylabel('Y', fontsize = 12)
plt.grid(True)
plt.legend()
plt.xlim(0, 6)
plt.ylim(0, 30)
plt.savefig('flashy_plot.png')
plt.show()
```

```python
data1 = np.random.normal(loc=0, scale=2, size=1000)
data2 = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data1, bins=20, color='blue', edgecolor='black')
plt.hist(data2, bins=20, color='pink', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.savefig('histogram.png')
plt.show()
```

```python
x = np.linspace(-3, 3, 1000)
y = np.linspace(-3, 3, 1000)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contourf(X, Y, Z, levels=20, cmap='viridis')  # Filled contours
plt.colorbar(label='Z Value')  # Add color legend
plt.title('Contour Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig('contour_plot.png')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')
plt.savefig('3d_surface_plot.png')
plt.show()
```

```python
x = np.linspace(0, 10, 100)

fig, axs = plt.subplots(2, 1, figsize=(4, 8))  # 1 row, 2 columns

# First plot: sine wave
axs[0].plot(x, np.sin(x))
axs[0].set_title('Sine Wave')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')

# Second plot: cosine wave
axs[1].plot(x, np.cos(x), 'r')
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')

plt.tight_layout()  # Adjust spacing
plt.savefig('subplots_example.png')
plt.show()
```

```python
from scipy.optimize import curve_fit

def model(x, a, b):
    return a * x + b

xdata = np.array([0, 1, 2, 3, 4])
ydata = np.array([1.0, 2.5, 5.5, 6.8, 9.1])  # Manually varied points

params, cov = curve_fit(model, xdata, ydata)

print(f"Fitted parameters: a={params[0]:.2f}, b={params[1]:.2f}")

plt.scatter(xdata, ydata)
plt.plot(xdata, model(xdata, *params), 'r-')
plt.savefig('curve_fit_example.png')
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def dydt(y, t): return -y

t = np.linspace(0, 5, 100)
y0 = 1
y = odeint(dydt, y0, t)

plt.plot(t, y)
plt.xlabel('Time'); plt.ylabel('y(t)')
plt.savefig('odeint_example.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

x = [0, 1, 2, 3]
y = [0, 1, 4, 9]

f = interp1d(x, y, kind='cubic')
x_new = np.linspace(0, 3, 100)

plt.plot(x, y, 'o', label='Data')
plt.plot(x_new, f(x_new), '-', label='Interpolant')
plt.legend()
plt.savefig('interpolation_example.png')
plt.show()
```

```python
from astropy import units as u

distance = 10 * u.meter
time = 5 * u.second

speed = distance / time
print(speed)  # Output -> 2.0 m / s

print(speed.to(u.km / u.hour))  # Convert units , Output -> 7.2 km / h

print((3 * u.meter) * (4 * u.meter)) # Combine units for area, Output -> 12 m2
```

```python
from astropy.time import Time

# Create a Time object in UTC
t = Time('2025-07-30 12:00:00', scale='utc')

# Convert to Julian Date (JD)
print("Julian Date:", t.jd)

# Convert to Modified Julian Date (MJD)
print("Modified Julian Date:", t.mjd)
```

```python
from astropy.coordinates import SkyCoord, AltAz, EarthLocation

loc = EarthLocation(lat=51.48*u.deg, lon=0*u.deg) # Observer location
time = Time('2025-07-30 22:00:00') # Time of observation
altaz = AltAz(alt=45*u.deg, az=130*u.deg, location=loc, obstime=time)
c = SkyCoord(altaz).icrs # Conversion

print(f"RA: {c.ra.degree:.2f}, Dec: {c.dec.degree:.2f}")
```

```python
from astropy.cosmology import Planck18 as cosmo

z = 1.5  # redshift

# Calculate distances at redshift z
dl = cosmo.luminosity_distance(z)
da = cosmo.angular_diameter_distance(z)
age = cosmo.age(z)

print(f"Luminosity distance: {dl:.2f}")
print(f"Angular diameter distance: {da:.2f}")
print(f"Age of Universe at z={z}: {age:.2f}")
```

```python
mass = 1.0          # kg
spring_constant = 10.0  # N/m
position = 0.1      # meters
velocity = 0.0      # m/s
```

```python
def update_velocity(dt):
    """
    Update the velocity of the mass after time step dt using SHM formula.
    
    Parameters:
        dt (float): time step in seconds.
    """
    global velocity
    acceleration = -(spring_constant / mass) * position
    velocity = velocity + acceleration * dt
    print(f"Velocity after {dt} s: {velocity:.3f} m/s")
```

```python
update_velocity(0.05)
```
