### `if` `elif` and `else`
We can combine conditional statements with `if` `elif` (or `else if`) and `else` to run *blocks* of code selectivley depending on if a statement evaluates to `True` or `False`. `elif`/`else` only run if the condition is `True`, and all above `if`/`elif` were `False`. Any `elif` and `else` statements after this will not run regardless. A new `if` statement will start this cycle again.

**Important**: Blocks of code are signified with indents - these indents are important as anything not indented in your block will not run with that block.

````python
x = 5

if x == 6:
    print('x is 6') # This would not run as 5 == 6 is False
elif (x > 3 or x == 7):
    print('x is larger than 3, or is 7') # This would run as 5 > 3 is True 
elif x == 5:
    print('x is 5') # This would not run as the clause above was already fulfilled
else:
    print('I give up') # This would not run as one of the above was True

if x==5:
    print('x is 5') # This is a new if statement, and it is True so it runs
````

````python
number = 1

while number <= 5:
  print(f"Number is: {number}")
    
  if number == 3:
    print("Stopping the loop early!")
    break  # exit the loop completely immediately
    print("That was a break!") # this will never run!
    
  number += 1 # pay attention to indents
  # this runs outside of the if, but inside the while
````

### Functions

Functions are like a mini-program inside your code, they take input arguments and return output arguments. Functions can be used to reuse code without copy/pasting chunks. We define functions with `def`, the contents must be indented, and values may (not must) be returned with `return`.

````python
def greet(name): # def function_name(parameter1, parameter2)
    return f"Hello, {name}!" # return output_parameters

print(greet("Natalie"))
print(greet("Nobody"))
````

Functions can take a number of arguments of varying data types. When passing arguments to functions in this way, they must be in the correct order, these are called *positional arguments*.

````python
def describe_planet(name, moons, mass):
    return f"{name} has {moons} moon(s) and a mass of {mass:.2e} kg."

print(describe_planet("Mars", 2, 6.39e23)) 
````

````python
def describe_star(name="Sun", temperature=5778):
    return f"{name} has a surface temperature of {temperature} K."

# Example usages
print(describe_star())                      # Uses both default values

print(describe_star("Sirius"))              # Overrides name, uses default temperature

print(describe_star("Betelgeuse", 3500))    # Overrides both
````

#### Example: Spacecraft Trajectory Simulation under Earth's Gravity

We simulate the 2D motion of a spacecraft orbiting Earth at 400 km altitude. The spacecraft’s position $\mathbf{r}(t)$ and velocity $\mathbf{v}(t)$ change due to Earth’s gravity:

The acceleration is

$$
\mathbf{a}(t) = - \frac{GM_{\mathrm{earth}}}{|\mathbf{r}(t)|^3} \mathbf{r}(t),
$$

and the velocity and position update by

$$
\mathbf{v}(t+\Delta t) = \mathbf{v}(t) + \mathbf{a}(t) \Delta t, \quad
\mathbf{r}(t+\Delta t) = \mathbf{r}(t) + \mathbf{v}(t+\Delta t) \Delta t.
$$

We want to simulate and analyze the trajectory over 10 minutes.


````python
# Constants
G = 6.67430e-11        # gravitational constant (m^3/kg/s^2)
M_earth = 5.972e24     # Earth mass (kg)
R_earth = 6.371e6      # Earth radius (m)

# Time array: 0 to 600 seconds, 61 points (10 minutes, 10s step)
times = np.linspace(0, 600, 61)

# Initial conditions
pos = np.array([R_earth + 400e3, 0.0])  # initial position (m)
vel = np.array([0.0, 7670.0])           # initial velocity (m/s)

positions = np.zeros((len(times), 2))   # to store position vectors

for i, t in enumerate(times):
    r = np.sqrt(pos[0]**2 + pos[1]**2)
    acc = -G * M_earth / r**3 * pos

    vel = vel + acc * (times[1] - times[0])
    pos = pos + vel * (times[1] - times[0])

    positions[i] = pos

# Analyze:
final_pos = positions[-1]
speeds = np.sqrt(positions[:,0]**2 + positions[:,1]**2)  # radial distances (m)
avg_distance = np.mean(speeds)

flight_angle = np.arctan2(vel[1], vel[0]) * 180 / np.pi  # degrees

print(f"Final position (km): {final_pos / 1000}")
print(f"Average distance from Earth center (km): {avg_distance / 1000:.1f}")
print(f"Final flight angle (deg): {flight_angle:.2f}")
print(f"Position at t=300s (km): {positions[30] / 1000}")
````

#### Example: Monte Carlo Estimation of \(\pi\)

Here we use Monte Carlo method to estimate the value of $\pi$ by randomly sampling points inside a square and checking how many fall within the inscribed unit circle.

We generate $N$ random points $(x, y)$ uniformly distributed in the square $[-1, 1] \times [-1, 1]$. The ratio of points inside the circle defined by

$$
x^2 + y^2 \leq 1
$$

approximates the ratio of the areas of the circle to the square. Since the square has area 4 and the circle area $\pi$, we estimate $\pi$ as:

$$
\pi \approx 4 \times \frac{\text{Number of points inside the circle}}{N}
$$

````python
# Generate 1,000,000 random points in the square [-1,1] x [-1,1]
n_points = 1_000_000
x = np.random.uniform(-1, 1, n_points)
y = np.random.uniform(-1, 1, n_points)

# Compute distance from origin for each point
dist = np.sqrt(x**2 + y**2)

# Count how many points fall inside the unit circle (distance <= 1)
inside_circle = dist <= 1

# Estimate pi as 4 * (fraction of points inside circle)
pi_estimate = 4 * np.mean(inside_circle)

print(f"Estimated π = {pi_estimate:.6f}")
````

#### Example: Two observatories located at different points on Earth observe the same asteroid. Each observatory measures the angle between their baseline and the line of sight to the asteroid.

Using trigonometry, the position $(x,y)$ relative to observatory 1 of the asteroid satisfies two linear equations derived from the angles observed:

$$
\begin{cases}
y = x \tan(\theta_1) \\
y = (x - d) \tan(\theta_2)
\end{cases}
$$

where $\theta_1, \theta_2$ are the observed angles at observatories 1 and 2, and $d$ is the distance between the two observatories along the x-axis (baseline).

These equations become:

$$
\begin{cases}
- \tan(\theta_1) x + y = 0 \\
- \tan(\theta_2) x + y = - d \tan(\theta_2)
\end{cases}
$$

or in matrix form


$$
\begin{bmatrix}
-\tan(\theta_1) & 1 \\
-\tan(\theta_2) & 1
\end{bmatrix}
\begin{bmatrix}
x \\ y
\end{bmatrix}
=
\begin{bmatrix}
0 \\ -d \tan(\theta_2)
\end{bmatrix}
$$

We observe $\theta_1 = 30^\circ, \theta_1 = 60^\circ$ and we know the distance between observatories as $d=100$ km. We can use `np.linalg.solve` to solve for $(x,y)$

````python
theta1 = np.radians(30)
theta2 = np.radians(60)
d = 100  # km between observatories

A = np.array([[-np.tan(theta1), 1],
              [-np.tan(theta2), 1]])
b = np.array([0, -d * np.tan(theta2)])

x, y = np.linalg.solve(A, b)
print(f"x = {x:.2f} km, y = {y:.2f} km")
````

#### Example: Spectral Analysis of a Galaxy

In this example, we analyze the spectral flux data of a galaxy measured at different wavelengths (in nanometers) from some telescope. The flux values represent the light intensity received at each wavelength.

The dataset contains pairs of wavelength $\lambda$ and flux $F(\lambda)$. We compute some statisitcs of the data.


````python
# Load data (skip header with information)
data = np.loadtxt('galaxy_spectrum.csv', delimiter=',', skiprows=1)

# Slice columns
wavelength = data[:, 0]  # in nm
flux = data[:, 1]        # arbitrary units

# Calculate statistics on flux
mean_flux = np.mean(flux)
var_flux = np.var(flux)
std_flux = np.std(flux)

# Find wavelength with max flux (peak emission)
peak_index = np.argmax(flux)
peak_wavelength = wavelength[peak_index]

print(f"Mean flux: {mean_flux:.2e}")
print(f"Flux variance: {var_flux:.2e}")
print(f"Flux standard deviation: {std_flux:.2e}")
print(f"Peak emission at wavelength: {peak_wavelength} nm with flux {flux[peak_index]:.2e}")
````

````python
# Generate x-values (log-spaced)
x = np.logspace(0.1, 2, 100)

# Line data: ideal model
y_line = 1e-2 * x**2

# Scatter data: simulated noisy measurements
noise = np.random.normal(loc=1.0, scale=0.2, size=len(x))
y_scatter = y_line * noise

# Create the figure
plt.figure(figsize=(8, 6))

# Plot the line
plt.plot(x, y_line, label="Model: $y = 10^{-2} x^2$", color="navy")

# Plot the scatter points
plt.scatter(x, y_scatter, label="Observed Data", color="orange", alpha=0.7) # alpha sets the opacity

# Set log scale
plt.xscale('log')
plt.yscale('log')

# Add labels, legend, and title
plt.xlabel("X (log scale)")
plt.ylabel("Y (log scale)")
plt.title("Log-Log Plot: Model vs Observed Data")
plt.legend()

# Show the plot
plt.show()
````

## Image visualisation

We can also use `matplotlib` fo image visualisation though `imshow` (though [`astropy`](https://docs.astropy.org/en/stable/) may be better suited depending on your problem)

````python
# Create grid
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y) # meshgrid takes the x and y arrays, and creates a new array of the complete set of [x , y] points

# Parameters for an elliptical galaxy-like brightness profile
x0, y0 = 0, 0      # Center
a, b = 2.0, 1.0    # Major and minor axis scale lengths
theta = np.deg2rad(30)  # Rotation angle

# Rotate coordinate system
X_rot = (X - x0) * np.cos(theta) + (Y - y0) * np.sin(theta)
Y_rot = -(X - x0) * np.sin(theta) + (Y - y0) * np.cos(theta)

# 2D elliptical Gaussian profile
brightness = np.exp(-(X_rot**2 / a**2 + Y_rot**2 / b**2))

# Add observational noise
noise = 0.05 * np.random.randn(*brightness.shape)
image = brightness + noise

# Plot
plt.figure(figsize=(6, 6))
plt.imshow(image, origin='lower', extent=[-5, 5, -5, 5], cmap='plasma')
plt.colorbar(label="Surface Brightness")
plt.title("Simulated Galaxy Brightness Map")
plt.xlabel("Arcsec")
plt.ylabel("Arcsec")
plt.grid(False)
plt.show()
````

## Subplots

Subplots allow multiple plots to be put in the same figure. This is usually done by defining axes that can then be indexed. Subplots assumes an equal split between plots, if you want some subplots to take up more space, you can use [`gridspec`](https://matplotlib.org/stable/api/_as_gen/matplotlib.gridspec.GridSpec.html) (though we won't cover gridspec here). # `scipy`

`scipy` is an extension of `numpy`, which introduces more sophisticated tools for numerical methods. We will go through some of the most used ones here.

## Optimisation and root finding: `scipy.optimize`

Here we go though some of the common functions used to optimise and root solve in Python

### Minimising functions

The function `minimize` can be used to minimise a chosen function given an initial guess:

````python
from scipy.optimize import minimize

# Define a function
def f(x, a, b):
    return a * (x - b)**2 + 5

# Parameters
a = 2
b = 3

# Initial guess
x0 = 0

# Pass extra args as a tuple after x0
result = minimize(f, x0, args=(a, b))

print("Minimum at:", result.x)
print("Minimum function value:", result.fun)
````

### Root solving

In very similar syntax we can find the root of functions:

````python
from scipy.optimize import root

# Define the function whose root we want
def g(x):
    return x**2 - 9

# Initial guess
x0 = 2

# Call root solver
sol = root(g, x0)

print("Root found at:", sol.x)
print("Function value at root:", g(sol.x))
````

### Fitting functions

Given a function with unknown parameter values, we can fit data to the function with `curve_fit`. In this example we fit for parameters $\{a, b, c\}$ in a Gaussian:
$$
y = a \exp\left(-\frac{(x - b)^2}{2 c^2}\right)
$$

````python
from scipy.optimize import curve_fit

# Define Gaussian model
def gaussian(x, a, b, c):
    return a * np.exp(-((x - b)**2) / (2 * c**2))

# Generate synthetic data
xdata = np.linspace(-5, 5, 100)
y_true = gaussian(xdata, 3, 0, 1)
noise = 0.2 * np.random.normal(size=xdata.size)
ydata = y_true + noise

# Fit curve
popt, pcov = curve_fit(gaussian, xdata, ydata, p0=[1, 0, 1])

print("Fitted parameters:")
print("a =", popt[0])
print("b =", popt[1])
print("c =", popt[2])

# Plot data and fit
plt.scatter(xdata, ydata, label='Data', s=15)
plt.plot(xdata, gaussian(xdata, *popt), 'r-', label='Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
````

## Integration and ODE solving: `scipy.integrate`

### Definite integration on a continous function

Function can be integrated with bounds applied using `quad`:

````python
from scipy.integrate import quad

# Define the function to integrate
def f(x):
    return np.sin(x)

# Integrate from 0 to pi
result, error = quad(f, 0, np.pi)

print(f"Integral result: {result}")
print(f"Estimated error: {error}")
````

### Definite integration on discrete data

If we want to integrate given some datasst, we can use `trapezoid` (for the trapezium rule, or we could use `simps` for Simpson's ruke):

````python
from scipy.integrate import trapezoid

# Generate discrete data points
x = np.linspace(0, np.pi, 50)
y = np.sin(x) + 0.1 * np.random.normal(size=x.size)  # sine with noise

# Compute integral using trapezoidal rule
integral = trapezoid(y, x)

print(f"Approximate integral (trapz): {integral}")

# Plot the data points
plt.scatter(x, y, label='Data points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Discrete noisy sine data')
plt.legend()
plt.show()
````

### ODE solving

We can use the function `odeint` to solve ODEs, here we show an example with a pendulum:

The equation of motion for a simple pendulum of length \( L \) and gravitational acceleration \( g \) is given by:

$$
\frac{d^2 \theta}{dt^2} + \frac{g}{L} \sin(\theta) = 0
$$

We rewrite this as a system of first-order ODEs:

$$
\begin{cases}
\frac{d\theta}{dt} = \omega \\
\frac{d\omega}{dt} = -\frac{g}{L} \sin(\theta)
\end{cases}
$$

where $ \theta $ is the angular displacement and $ \omega $ is the angular velocity.

We solve this system numerically and plot $\theta(t) $.


````python
from scipy.integrate import odeint

# Parameters
g = 9.81  # gravity (m/s^2)
L = 1.0   # pendulum length (m)

# Define the system of ODEs
def pendulum(y, t):
    theta, omega = y
    dtheta_dt = omega
    domega_dt = -(g / L) * np.sin(theta)
    return [dtheta_dt, domega_dt]

# Initial conditions: theta=45 degrees, omega=0
theta0 = np.radians(45)
omega0 = 0
y0 = [theta0, omega0]

# Time points where solution is computed
t = np.linspace(0, 10, 300)

# Solve ODE
sol = odeint(pendulum, y0, t)

# Plot results
plt.plot(t, np.degrees(sol[:, 0]), label=r'$\theta(t)$')
plt.xlabel('Time [s]')
plt.ylabel('Angle [degrees]')
plt.title('Simple Pendulum Motion')
plt.legend()
plt.grid(True)
plt.show()
````

## Interpolation

One dimensional interpolation can be acheived with `interp1d` for a discrete set of $(x, y)$ values, with a linear interpolation used by default or cubic can be specified:

````python
from scipy.interpolate import interp1d

# Sample data points
x = np.linspace(0, 10, 10)
y = np.sin(x)

# Create interpolation functions
linear_interp = interp1d(x, y,)
cubic_interp = interp1d(x, y, kind='cubic')

# New points for evaluation
x_new = np.linspace(0, 10, 200)

# Evaluate interpolations
y_linear = linear_interp(x_new)
y_cubic = cubic_interp(x_new)

# Plotting
plt.scatter(x, y, label='Data points', color='black')
plt.plot(x_new, y_linear, label='Linear interp', linestyle='--')
plt.plot(x_new, y_cubic, label='Cubic interp1d')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of Interpolation Methods')
plt.grid(True)
plt.show()

````

## Constants

`scipy` has a host of constants we can use without coding them up, the full list is [here](https://docs.scipy.org/doc/scipy/reference/constants.html)

````python
from scipy.constants import h, c, e  # Planck constant, speed of light, elementary charge

# Given wavelength in nanometers
wavelength_nm = 500  # green light

# Convert wavelength to meters
wavelength_m = wavelength_nm * 1e-9

# Calculate photon energy in joules: E = h * c / lambda
energy_joules = h * c / wavelength_m

# Convert energy to electron volts
energy_eV = energy_joules / e

print(f"Photon energy at {wavelength_nm} nm: {energy_joules:.2e} J = {energy_eV:.2f} eV")
````
